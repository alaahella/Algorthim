      
       **Using Networkx to draw Graph**

01.PNG

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
G.add_edges_from([(0,1),(0,8),(0,3)])
G.add_edges_from([(1,0),(1,7)])
G.add_edges_from([(3,0),(3,2),(3,4)])
G.add_edges_from([(4,3),(4,8)])
G.add_edges_from([(5,2),(5,6)])
G.add_edges_from([(2,3),(2,5),(2,7)])
G.add_edges_from([(7,1),(7,2)])
G.add_edges_from([(6,5)])
G.add_edges_from([(8,0),(8,4)])
nx.draw(G,with_labels=True,node_size=1500,node_color=["green",'yellow','yellow','yellow','yellow','yellow','yellow','yellow','yellow'])
plt.show()
```

       
       **DFS path for the Graph**

```python
Dfs=nx.depth_first_search.dfs_tree(G,0)
nx.draw(Dfs,with_labels=True,node_size=1500,node_color=["green",'yellow','yellow','yellow','yellow','yellow','yellow','yellow','yellow'])
```

      
      **BFS Path for the graph**

```python
Bfs=nx.breadth_first_search.bfs_tree(G,0)
nx.draw(Bfs,with_labels=True,node_size=1500,node_color=["green",'yellow','yellow','yellow','yellow','yellow','yellow','yellow','yellow'])
```

        
        **Dijkstra for the following weighted graph**

02.png

```python
G=nx.Graph()
G.add_edge(0, 1, weight=8)
G.add_edge(0, 2, weight=9)
G.add_edge(1, 3, weight=1)
G.add_edge(1, 5, weight=7)
G.add_edge(1, 6, weight=9)
G.add_edge(2, 5, weight=6)
G.add_edge(2, 6, weight=4)
G.add_edge(3, 7, weight=4)
G.add_edge(5, 7, weight=2)
G.add_edge(7, 6, weight=6)
G.add_edge(6, 4, weight=7)
G.add_edge(4, 7, weight=7)
pos= nx.spring_layout(G)
nx.draw_networkx(G,pos,with_labels=True,node_size=1500,node_color=["green",'yellow','yellow','yellow','yellow','yellow','yellow','yellow'])
nx.draw_networkx_edge_labels(G,pos,edge_labels=nx.get_edge_attributes(G ,'weight' ))
plt.show()
```

    
      **the shortest path**

```python
Dijkstra=nx.dijkstra_path_length(G,0,7)
print("the shortest path =",Dijkstra)
```

```python
Dijkstra=nx.dijkstra_path(G,0,7)
Dijkstra
```
