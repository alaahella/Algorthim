```python
def dfs(visited, graph, node):
    if node not in visited:
        visited.append(node)
        for neigbour in graph[node]:
            dfs(visited, graph, neigbour)

graph = { "0": ["1", "3", "8"],
          "1": ["0", "7"],
          "2": ["3", "5", "7"],
          "3": ["0", "4","8"],
          "4": ["3", "8"],
          "5": ["2", "6"],
          "6": ["5"],
          "7": ["1", "2"],
          "8": ["0", "4"]}

visited = []
dfs(visited, graph, "0")
print("Depth-First Search:",visited)
```

      ** BFS Path for the graph**

```python
def bfs(visited, graph, node): 
    visited.append(node)
    queue.append(node)
    while queue:          
        for neighbour in graph[queue.pop(0)]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

visited = [] 
queue = []      
bfs(visited, graph, '0')
print("Breadth-First Search:",visited)
```

```python
def dijsktras(graph,startNode,endNode):
    if(startNode in graph and endNode in graph ):
        shortestPath = {}
        shortestPathWeight ={}
        for keys in graph.keys():
            shortestPathWeight[keys]=float("inf")
            shortestPath[keys] = []
        shortestPathWeight[startNode]=0
        myQueue = []
      
        for items in graph[startNode].keys():
            myQueue.append(items)
            shortestPath[items].append(startNode)
            shortestPath[items].append(items)
            shortestPathWeight[items] = graph[startNode][items]
        while( len(myQueue)>0):
            currrentKey =myQueue.pop(0)
            for items,weights in graph[currrentKey].items():
                if(weights+shortestPathWeight[currrentKey]<shortestPathWeight[items]):
                    myQueue.append(items)
                    shortestPath[items]=[]
                    shortestPathWeight[items] = weights+shortestPathWeight[currrentKey]
                    shortestPath[items].extend(shortestPath[currrentKey])
                    shortestPath[items].append(items)
                    
        return shortestPathWeight[endNode], shortestPath[endNode]
    else:
        return

graph = {
                    0: {1:8, 2:9},
                    1: {0:8, 5:7, 3:1, 6:9},
                    2: {0:9, 5:6, 6:4},
                    3: {1:1, 7:4},
                    4: {7:7, 6:7},
                    5: {1:7, 2:6, 7:2},
                    6: {1:9, 2:4, 4:7, 7:6},
                    7: {3:4, 4:7, 5:2, 6:6}
                }
distance, path = dijsktras(graph, 0, 7)
print("Shortest path from 0 to 7 is:  {}".format(path))
print("The weight of shortest path = {}".format(distance))
```
